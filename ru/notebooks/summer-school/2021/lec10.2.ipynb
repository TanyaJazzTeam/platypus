# Возможности и возможности моделей квантового машинного обучения и будущее квантового машинного обучения

В этой лекции Амира представляет различные открытые вопросы, касающиеся квантового машинного обучения и квантовых вычислений в целом. Цель состоит в том, чтобы объяснить потенциал и мощь квантового машинного обучения для приложений сегодня и завтра. Она начинает с того, что фокусируется на определении емкости в области классического машинного обучения. Действительно, в нем много определений: статистическая сложность, выразительность, мощность. Это можно резюмировать, задаваясь вопросом, сколько функций может аппроксимировать модель. Чем больше он может аппроксимировать, тем больше у него возможностей. Однако, как это ни парадоксально, большая емкость не обязательно является хорошей вещью. Емкость связана с обобщением, связанным с компромиссом между смещением и дисперсией, когда мы не хотим, чтобы модель не подгоняла и не подгоняла. Таким образом, оптимальная пропускная способность — это модель с наименьшей ошибкой обобщения в компромиссе смещения/дисперсии. Таким образом, вопрос о том, как измерить мощность модели машинного обучения как в классических, так и в квантовых вычислениях, очень важен и открыт.
