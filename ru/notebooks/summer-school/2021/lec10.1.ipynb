# Расширенные алгоритмы QML: квантовые машины Больцмана и квантово-генеративные состязательные сети

В этой лекции Криста представляет алгоритмы квантового машинного обучения, применяемые для подготовки состояния. Мотивация этой темы заключается в необходимости найти приблизительный, но эффективный способ кодирования данных в квантовые состояния, чтобы исследовать квантовые алгоритмы, которые могли бы обеспечить вычислительное преимущество без экспоненциально дорогой нагрузки. Здесь мы хотим использовать параметризованные квантовые схемы и методы машинного обучения. Таким образом, Криста представляет два алгоритма, которые решают эту проблему, сосредоточив внимание на загрузке распределения вероятностей. Первая — это квантовые генеративные состязательные сети или QGAN, напрямую взятые из классического машинного обучения, где генератор должен генерировать данные, которые будут обрабатываться дискриминатором. Способность дискриминатора распознавать, являются ли данные сгенерированными или исходными, напрямую влияет на обучение как генератора, так и самого дискриминатора. Второй — квантовые машины Больцмана, название которых основано на функции вероятности, полученной из термодинамики с постоянной Больцмана. Действительно, обучающая часть обновит эту функцию вероятности, чтобы она могла представлять базовое распределение вероятностей набора данных.
