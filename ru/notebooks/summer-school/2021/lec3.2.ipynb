В этой лекции Златко продолжает тему шума в квантовых вычислениях, уделяя особое внимание выборке, СПАМу и некогерентным ошибкам. Златко обсуждает теорию квантовых измерений на примере измерения фон Неймана и влияния выборочных результатов измерений. Он связывает это с тем, как измерения, точнее, ошибки измерения и выборки влияют на алгоритмическую производительность квантовых компьютеров. Затем он переходит к ошибкам SPAM (состояние подготовки и измерения) и обсуждает, как они влияют на производительность схемы. Наконец, он заканчивает лекцию кратким описанием бессвязных ошибок.

### Часто задаваемые вопросы

<details>   <summary>What exactly do the gates manipulate on the qubit when applying the gates? In the slides we represent that manipulation as the arrow moving, but in the physical world, what does that translate to?</summary> The black arrow that Zlatko showed in some of the pictures (See, slides: 16-23) represents the direction along which the gate is acting, in other words the normal to the plan of rotation—it’s the vector associated with the rotation that is caused. The colored arrows that he showed depicting the trajectory of the state factor along the surface of the blocks fear or depictions of the evolution of the wavefunction as a function of time under the application of the gate, considered not as an instantaneous, but it’s a real finite in time evolution. We idealize the gates to be instantaneous in a quantum algorithm, but in practice they always have some finite time, and the state doesn’t just jump, it must take a continuous pass through Hilbert space. </details>


<summary>Это когерентный шум, вызванный постоянной ошибкой в X-вороте? Например, если в вентиле X есть как положительные, так и отрицательные ошибки, ошибки полностью компенсируют шум. Итак, что вызывает ошибку в каждом X-гейте?</summary> Здесь мы предположили, что шум был детерминированным, так что каждый раз, когда вы применяете его к гейту, вы получаете тот же самый шум, а не другой. Это означает, что из-за когерентной природы шума каждое применение шума нарастает. Квадратичная ошибка первого порядка в худшем случае. Если бы шум колебался, то вы бы получили более медленное нарастание, потому что каждый раз, когда вы играли в гейт, вы иногда перемещались бы влево, иногда вправо, вы все равно дрейфовали бы в среднем, но медленнее.

<details>   <summary>If the error is average +2°,  for example: can we do an R_x(180°-2°) like correction, the application of a rotation gate would have an associated error with itself. In the lecture Zlatko said “a measurement of error would allow us to correct it”. Is the correction perhaps done in some manner other than the application of another gate?</summary> Right so the gate you would apply is the noisy \tilde{X} = R_x(182°), you then dial back the amplitude on the pulse by roughly a fraction 2/182 = 0.011, and check the gate with this error amplifying sequences. Then your gate should be roughly \tilde{X} = R_x(182° * 180/182) = R_x(180° ) </details>


<summary>«Корреляция измерений в состоянии Белла сильнее, чем когда-либо могла быть в классической системе» — что означает «более сильная» корреляция? Можем ли мы количественно определить корреляцию?</summary> Проверьте неравенства нарушения колокола.


<summary>Являются ли все проективные измерения измерениями фон Неймана? По измерениям фон Неймана, в основном при небольшом чтении по открытым системам, есть ли какие-либо ключевые моменты или предположения для обоих?</summary> Измерения фон Неймана являются наиболее ограниченным классом измерений. Обычно проективное и фон Неймановское означают одно и то же. Однако более тонко некоторые из нас более тщательно проводят различие между термином «измерение фон Неймана» и термином «проективные измерения», потому что фон Нейман неправильно понял его для проекторов ранга &gt; 1. Представьте себе оператор с вырожденными собственными значениями.


<summary>В лекции Златко показал «модель зашумленного вентиля с ошибкой битового перехода» (см. изображение ниже). Является ли этот вентиль своего рода «смесью» двух меньших схем, которые случайным образом выводят нули или единицы?</summary> Вот так. Это смесь различных схем следующим образом: когда вы запускаете схему, вы не знаете, какую из двух вы на самом деле будете запускать. Верхняя строка (с I) или нижняя строка (с X). Вы не можете сказать разницу. Таким образом, вы должны усреднить их. Это недостаток контроля и знаний.


<summary>При каких условиях рассчитывается ожидание оператора, использующего Trace?</summary> Когда используются матрицы плотности.
