### Вопросы-Ответы

<details><summary>Что означает, что состояние максимально запутано?</summary> Максимально запутанное состояние — это квантовое состояние (например, состояние Белла), которое имеет максимальную энтропию фон Неймана; которая аналогична классической энтропии Шеннона для подбрасывания честной монеты.</details>

<details><summary>Для квантовой телепортации нам также нужно отправлять классические биты, и мы не можем просто использовать запутанность для отправки информации. Не по этой ли причине мы не можем передавать информацию со скоростью, превышающей скорость света?</summary> да. На самом деле, поскольку само запутанное состояние представляет собой не распределение вероятностей, а его внутренний продукт, всегда приходится измерять состояние, чтобы извлечь из него информацию.</details>

<details><summary>Почему вы использовали представление внешнего произведения при применении вентиля вместо использования спектрального разложения?</summary> Представление внешнего произведения в терминах состояний 0 и 1 (так называемая вычислительная основа) полезно для вентиля, потому что тогда вы легко можете понять, как вентиль действует на состояния 0 и 1.</details>

<details><summary>Чем телепортация полезна в квантовых вычислениях по сравнению с классическими?</summary> Это полезно, когда вы хотите зашифровать данные (информацию), которые вы передаете (т. е. вместо того, чтобы клонировать их, вы фактически передаете их целевой стороне, что будет включать уничтожение исходной информации (данных) из источника).</details>
