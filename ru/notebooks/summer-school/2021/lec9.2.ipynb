### Часто задаваемые вопросы

<details><summary>Есть ли ссылки, где непрерывный набор ворот использовался для вариационных форм? Может ли оптимизатор настроить параметры пульса?</summary> Да! Мы проверили это с помощью QAOA, одного из типов вариационной формы. И я также начал тестировать это на схемах типа QML. Оптимизатор может настроить параметр импульса — вы бы хотели использовать импульсные ворота для бесшовной интеграции с модулями приложения (некоторые ранние тесты показали, что на самом деле это работает хорошо)</details>

<details><summary>И как транспилятор справляется с шумом? (источник: вопросы и ответы в прямом эфире на 4:11)</summary> Это зависит от шума. В частности, учитывая, что C-NOT имеет очень плохую достоверность (значительно большую недостоверность по сравнению с другими), преобразование шума сведет к минимуму то, сколько вы фактически используете этот плохой вентиль C-NOT, предоставляя разные маршруты. Имейте в виду, что транспирация шума может не работать должным образом из-за устаревшего снимка, сделанного в результате измерения, что приводит к нежелательным результатам подавления шума.</details>

<details><summary>Проработка регулировки фазы?</summary> Для лабораторной системы отсчета в эксперименте Раби мы можем отрегулировать ось вращения, выполнив настройку фазы. Другими словами, фазовая гамма (в формуле «U=..» в примечании к лекции 9.2, слайд 12) диктует, с какой осью мы в конечном итоге будем работать.</details>

<details><summary>Чем колебания Раби отличаются от колебаний Рамсея?</summary> Осцилляция Раби: начните с кубита в основном состоянии. Затем мы применяем несколько импульсов (обычно используются гауссовые или прямоугольные импульсы). Есть два типа осцилляции Раби: осцилляции Раби времени и осцилляции Раби мощности.</details>

- При осцилляции времени-Раби МАКСИМАЛЬНАЯ амплитуда импульса такая же, но мы увеличиваем длительность импульса (увеличиваем площадь под кривой пульса). Площадь (прямоугольник в примере из лекции) под кривой импульса определяет скорость, с которой кубит вращается 0 → 1 → 0 → 1 → 0 → 1 ......, поэтому называется пи-импульсом.
- При силовых колебаниях Раби длительность импульса фиксирована, но его максимальная амплитуда увеличивается. Площадь под кривой в этом случае увеличивается по вертикали.
- Осцилляции Рамсери: примените импульс pi/2, чтобы кубит вращался из основного состояния в состояние на экваторе сферы Блоха (см. слайд 14 в конспекте лекции для визуализации). Затем подождите некоторое время → подайте второй импульс → подождите → третий импульс... и так далее. В результате мы получаем осцилляции, которые можно использовать для точной настройки частоты кубита и определения T2
